```python
import matplotlib.pyplot as plt
import numpy as np

a = np.array([0, 1, 2, 3])
b = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(a, b)

a = np.array([0, 1, 2, 3])                     # learning subgraph using numpy
b = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.bar(a, b)

a = np.array([0, 1, 2, 3])
b = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(a, b)

a = np.array([0, 1, 2, 3])
b = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.bar(a, b)
plt.title("SALES")

a = np.array([0, 1, 2, 3])
b = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.bar(a, b)

a = np.array([0, 1, 2, 3])
b = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(a, b)

plt.suptitle("learning subgraphs")

plt.show()


```
